```python
import os
from os.path import join, basename, expanduser
from glob import glob
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from conversions import dB_convert
```

```python
result_dir = expanduser(f'~/infrasound/results/attenuation/')
with open(join(result_dir, 'attenuation_v2.pkl'), 'rb') as f:
    res = pickle.load(f)
```

```python
t = pd.DataFrame()
for hs, values in res.iteritems():
    h1, h2 = hs.split('-')
    z = round(abs(float(h2)- float(h1)), 1)
    zs = np.ones_like(values)*z
    values = np.abs(values)
    plt.scatter(zs, values, color = 'blue', alpha = 0.8)
    t = pd.concat([t, pd.DataFrame.from_dict({str(z):[np.nanmean(values)]})])

for z, values in t.iteritems():
    for v in values[~np.isnan(values)]:
        plt.scatter(float(z), v, color = 'red')
```

```python
plt.subplots(figsize = (12,8))
t = pd.DataFrame()
for hs, values in res.iteritems():
    h1, h2 = hs.split('-')
    z = round(abs(float(h2)- float(h1)), 1)
    zs = np.ones_like(values)*z*100
    values = np.abs(values)
    plt.scatter(zs, values, color = 'blue', alpha = 0.5, marker = 'x')
    t = pd.concat([t, pd.DataFrame.from_dict({str(z):[np.nanmean(values)]})])

for i, (z, values) in enumerate(t.iteritems()):
    for v in values[~np.isnan(values)]:
        if i == 0:
            plt.scatter(100*float(z), v, color = 'red', label = 'Sensor Pair Average Attenuation')
        else:
            plt.scatter(100*float(z), v, color = 'red')

line = {}
for z, values in t.iteritems():
    # for v in values[~np.isnan(values)]:
    #     plt.scatter(float(z), v, color = 'red')
    dB_diff = dB_convert(np.nanmean(values))
    diff = np.nanmean(values)
    z = np.round(100*float(z), 0)
    # plt.scatter(z, diff, color = 'blue')
    line[z] = diff
plt.plot(line.keys(), line.values(), label = 'Average Attenuation Values')

plt.ylim(0, 0.0003)
plt.legend()

plt.ylabel('Power Attenuated [$Pa^{2}/s$]')
plt.xlabel('Snow Between Sensors [cm]')
plt.title('Coherent Signal Attenuation')
```
