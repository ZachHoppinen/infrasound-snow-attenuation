```python
import gemlog
import matplotlib.pyplot as plt
```

```python
low = gemlog.ims_noise(model = 'low', spectype = 'power', freq_min = 0.1, freq_max = .5)
med = gemlog.ims_noise(model = 'med', spectype = 'power', freq_min = 0.1, freq_max = .5)
high = gemlog.ims_noise(model = 'high', spectype = 'power', freq_min = 0.1, freq_max = .5)
```

```python
plt.plot(low['freqs'], low['spectrum'], label = 'low')
plt.plot(med['freqs'], med['spectrum'], label = 'med')
plt.plot(high['freqs'], high['spectrum'], label = 'high')
plt.legend()
plt.yscale('log')
```
